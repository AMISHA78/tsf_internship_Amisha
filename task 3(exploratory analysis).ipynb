```python
correlation=x.corr()
plt.figure(figsize=(10,7))
sb.heatmap(correlation,annot=True)
```

```python
#data visualization
sb.countplot(data = x, x = 'Category');
plt.xticks(rotation=60)
```

```python
base_color=sb.color_palette()[1]
gen_order=x['Category'].value_counts().index
sb.countplot(data=x,x='Category',order=gen_order)
```

```python
base_color=sb.color_palette()[7]
sb.countplot(data=x,x='Sub-Category')
plt.title("sub product category")
plt.xticks(rotation=90)
```

```python

sb.pairplot(x)
```

```python
g=sb.FacetGrid(data=x,col='Category')
g.map(plt.hist,"Quantity")
```

```python
plt.figure(figsize=(10,10))
x['Category'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

```python
x.groupby('Region')['Profit'].agg(sum).plot(kind='bar',color='yellow')
plt.title('profit in different region')
```

```python
sb.lineplot(x['Discount'],x['Profit'])
```

```python
x.groupby('Category')['Profit','Sales'].agg(sum).plot(kind='bar')
plt.title('profit and sales')
```
