```python
x['imonth']=x['imonth'].fillna(0)

df=x[['iyear','imonth','iday']].copy()
df.columns=['year','month','day']

df=df.applymap(str)

x['time']=pd.to_datetime(df[['year','month','day']],format='%Y%m%d',errors='coerce')
x.head(15)
```

```python
l=x.isnull().sum()
g=pd.DataFrame(l)
g
```

```python

g['features']=x.columns.values
g
```

```python
g['features']
```

```python

plt.figure(figsize=(20,20))
sb.barplot(x=g[0],y=g['features'])
```

```python
corr=x.corr()
plt.figure(figsize=(18,15))
sb.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns)
```

```python
col=['iyear','imonth','iday','time','country_txt','nkill','region','city','attacktype1_txt','nwoundte']
newdata=x[col]
newdata
```

```python
 naming={'iyear':'year','imonth':'month','country_txt':'country','nkill':'kill','nwoundte':'wound','attacktype1_txt':'attacktype'}
newdata.rename(columns=naming,inplace=True)
newdata
```

```python
fig,ax3=plt.subplots(figsize=(40,30))
newdata.hist(ax=ax3)
```

```python
df=newdata[['year','kill']].groupby(['year']).sum()
df

fig,ax4=plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
ax4.set_title("people who died due to covid")
plt.xlabel('Year')
plt.ylabel('Number of killed')
```

```python
pd.crosstab(newdata.year,newdata.region).plot(kind='area',figsize=(16,8))
plt.xticks(np.ceil(np.linspace(newdata['year'].min(),newdata['year'].max(),20)))
plt.title("terrorist activity by region")
plt.ylabel("number of attack")
```

```python
#countplot attack per year
plt.subplots(figsize=(15,10))
sb.countplot("year",data=newdata)
plt.xticks(rotation=90)
plt.title("attacks per year")
```

```python
#increase in attack from 1970 to 2017
a=newdata['year'].value_counts().to_dict()
x=np.round(((a[2017]-a[1970])/a[2017])*100)
print(f"the number of attacks increased since 1970 to 2017 is {x}")
```

```python
#top 10 cities  affected by attacks
newdata['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(15,10),color='yellow')
plt.title("top 10 affected cities")
plt.xlabel("city")
plt.ylabel("number of attack")
```

```python
newdata['attacktype'].value_counts().to_frame().sort_values('attacktype',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(15,10),color='yellow')
plt.title("top 10 affected cities")
plt.xlabel("attack type")
plt.ylabel("number of attack")
```
