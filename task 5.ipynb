```python
season_data=df[['id','season','winner']]

complete_data=df1.merge(season_data,how='inner',left_on='match_id',right_on='id')
complete_data
```

```python
four_run=complete_data[complete_data['batsman_runs']==4]
four_run.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
batsman_four=four_run.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
```

```python
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
batsman_score=df1.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```
